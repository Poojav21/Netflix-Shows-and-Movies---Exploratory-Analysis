```python
directors_count = df['director'].nunique()
print(f"Number of unique directors: {directors_count}")
directors_count = df['director'].value_counts()
print(f"Number of movies directed by each director:\n{directors_count}")
```

```python
df_filtered = df[df['director'] != 'Not Given']
directors_count = df_filtered['director'].value_counts()

# Display the top 10 directors
top_10_directors = directors_count.head(10)
print("Top 10 directors:\n", top_10_directors)
```

```python
# Create a bar chart using plotly
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
fig = go.Figure([go.Bar(x=top_10_directors.index, y=top_10_directors.values, marker_color = colors)])
fig.update_layout(
    title='Top 10 Directors by Number of Movies Directed',
    xaxis_title='Director',
    yaxis_title='Number of Movies Directed',
    xaxis_tickangle=-45
)
fig.show()
```

```python
df['country'].unique()
```

```python
country_counts = df['country'].value_counts()
country_counts
```

```python
top_10_directors = country_counts.head(10)
print("Top 10 directors:\n", top_10_directors)
```

```python
country_type_counts = df.groupby(['country', 'type']).size().reset_index(name='count')

# Get the top 10 countries by total number of entries
top_10_countries = df[df['country'] != 'Not Given']['country'].value_counts().nlargest(10).index
filtered_df = country_type_counts[country_type_counts['country'].isin(top_10_countries)]
filtered_df

```

```python
fig = px.bar(
    filtered_df,
    x='country',
    y='count',
    color='type',
    barmode='group',  # Ensures counts are displayed side by side
    title='Top 10 Countries by Number of Movies and TV Shows',
    color_discrete_sequence=px.colors.qualitative.Dark24  # Use a darker color scheme
)

fig.update_layout(
    xaxis_title='Country',
    yaxis_title='Count',
    legend_title='Type',
    xaxis_tickangle=-45
)

fig.show()
```

```python
top_year = df['release_year'].value_counts().reset_index()
top_year.columns = ['release_year', 'count']
top_year
```

```python
top_year = top_year.sort_values(by='count', ascending=True)


fig = px.scatter(
    top_year,
    x='release_year',
    y='count',
    color='release_year',
    title='Top Years in which Most Movies Released',
    labels={'release_year': 'Years', 'count': 'Numbers'},
    color_continuous_scale=px.colors.qualitative.Dark24
)


fig.update_layout(
    xaxis_tickangle=-45
)

fig.show()
```

```python
df = df[df['country'] != 'Not Given']


grouped_df = df.groupby(['release_year', 'country']).size().reset_index(name='count')


fig = px.line(
    grouped_df,
    x='release_year',
    y='count',
    color='country',
    title='Number of Movies and TV Shows by Country Over Years',
    labels={'release_year': 'Release Year', 'count': 'Number of Entries'}
)

fig.show()
```

```python
tv_shows = df[df['type'] == 'TV Show'].copy()
tv_shows['num_seasons'] = tv_shows['duration'].str.extract(r'(\d+)', expand=False).astype(float)


fig = px.histogram(
    tv_shows,
    x='num_seasons',
    nbins=10,
    title='Distribution of Seasons for TV Shows',
    labels={'num_seasons': 'Number of Seasons'},
    color_discrete_sequence=px.colors.qualitative.Dark24
)


fig.update_layout(
    xaxis_title='Number of Seasons',
    yaxis_title='Frequency',
    bargap=0.2,
    showlegend=False
)

fig.show()
```
